```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
```

```python
ourfreqs = np.loadtxt('../data/exc/1RWT.exc/freqs.txt')

litlambdas = np.array([635, 652, 647, 656, 630, 636, 629, 640, 634, 656, 656, 663, 646, 660])
litfreqs = (1.0e+7/litlambdas)

# 1RWT contains 10 "identical" chains, with 14 pigments each
ourfreqs = np.reshape(ourfreqs, (10,14))
meanfreqs = np.mean(ourfreqs, 0)
plt.plot([15000, 16000], [15000, 16000], 'k-')
plt.plot(litfreqs, meanfreqs, 'o')
plt.xlabel('Literature value (cm$^{-1}$)')
plt.ylabel('PigmentHunter value (cm$^{-1}$)')
plt.gca().set_aspect('equal')
plt.show()
```
