```python
import os,sys
sys.path.append('./misc/lib/python3.7/site-packages')

import math
import numpy as np
import requests
import nglview as nv
import ipywidgets as widgets
%matplotlib notebook
import matplotlib.pyplot as plt
from IPython.display import display, display_markdown
from ipywidgets import Layout, HTML
from pathlib import Path

import parmed as pmd
import re

import hublib.use
from hublib.ui import FileUpload, Download
from hublib.cmd import runCommand

from scipy.ndimage import gaussian_filter

import time

%matplotlib notebook
%use gromacs-2018.4
%use pymol-1.8.4

np.set_printoptions(precision=8)
np.set_printoptions(suppress=True)
```

```python
out = !{'cp ../pdb2gmx/complex.gro ./'}
out = !{'cp ../pdb2gmx/complex.top ./'}
out = !{'cp ../pdb2gmx/*.itp ./'}

MainAcc = widgets.Accordion(children=[widgets.Textarea(layout=Layout(width='15cm', height='10cm')),
                                      widgets.Textarea(layout=Layout(width='15cm', height='10cm')),
                                      widgets.Textarea(layout=Layout(width='15cm', height='10cm'))])
MainAcc.set_title(0, 'GROMPP')
MainAcc.set_title(1, 'MDRUN')
MainAcc.set_title(2, 'Potential Energy')
display(MainAcc)

result = ""
# Use editconf to set the box size. 
out = !{"gmx editconf -f complex.gro -o newbox.gro -bt cubic -d 1"}
for line in out:
    result += line + "\n"
result += "\n\n"

# Compile a *.tpr file for the energy minimization
out = !{"gmx grompp -maxwarn 1 -f em.mdp -c newbox.gro -p complex.top -o em.tpr"}
for line in out:
    result += line + "\n"
MainAcc.children[0].value = result
MainAcc.selected_index = 0

# Run the minimization
result = ''
out = !{"gmx mdrun -v -deffnm em"}
for line in out:
    result += line + "\n"
MainAcc.children[1].value = result
MainAcc.selected_index = 1

# Print the charges
out = !{"gmx editconf -f em.tpr -mead -o charge.pdb"}
out = !{"bash parse_pqr.sh mead.pqr"}

# Output a PDB file for viewing
out = !{"echo -e 0 | gmx trjconv -f em.gro -s em.tpr -o em.pdb"}

# Output potential curve for validation
result = ''
out = !{"echo -e 10 0 | gmx energy -f em.edr -o potential.xvg"}
for line in out:
    result += line + "\n"
MainAcc.children[2].value = result
MainAcc.selected_index = 2

display_markdown('# Potential Energy Curve #', raw=True)
display_markdown('Please check for consistency. You should see a smooth decay.', raw=True)
fig = plt.figure(figsize=(6,4))

dat = np.loadtxt('potential.xvg', comments=['@', '#'])
plt.plot(dat[:,0], dat[:,1]/1000.0)
plt.xlabel('Time (ps)', fontsize=16)
plt.ylabel('Potential Energy ($10^6$ J/mol)', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()
plt.show()
```
